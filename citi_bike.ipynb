```python
import pandas as pd
```

```python
#read in December 2023 CSV and Jan 2024 CSV 
dec_data= pd.read_csv("../tableau_challenge/resources/JC-202312-citibike-tripdata.csv")
jan_data= pd.read_csv('../tableau_challenge/resources/JC-202401-citibike-tripdata 2.csv')

#checking out the first 10 rows of Dec 2023 data
dec_data.head(10)
```

```python
#checking out the first 10 rows of Jan 2024 data 
jan_data.head(10)
```

```python
#printing length of each csv to see how many rows in Dec 2023 CSV and Jan 2024 CSV
print(len(dec_data))
print(len(jan_data))
```

```python
#dropping any null values and re-checking length to see how many rows dropped for Dec 2023 CSV
dec_dropna= dec_data.dropna()
len(dec_dropna)
```

```python
#dropping any null values and re-checking length to see how many rows dropped for Jan 2024 CSV
jan_dropna= jan_data.dropna()
len(jan_dropna)
```

```python
#dropping duplicate values and re-checking length for Dec 2023 csv, however no duplicates seemed to be found
dec_no_dupes= dec_dropna.drop_duplicates()
len(dec_no_dupes)
```

```python
#dropping duplicate values and re-checking length for Jan 2024 csv, however no duplicates seemed to be found
jan_no_dupes= jan_dropna.drop_duplicates()
len(jan_no_dupes)
```

```python
#saving cleaned Dec 2023 csv as a new csv to import into tableua
dec_no_dupes.to_csv('cleaned_dec_citi_bike.csv', index=False)
```

```python
#saving cleaned Jan 2024 csv as a new csv to import into tableua
jan_no_dupes.to_csv('cleaned_jan_citi_bike.csv', index=False)
```

```python
#reading back in and confirming length to make sure export was correct for both csv files
cleaned_dec= pd.read_csv("../tableau_challenge/resources/cleaned_dec_citi_bike.csv")
cleaned_dec.head()
# print(len(cleaned_dec))
```

```python
#reading back in and confirming length to make sure export was correct for both csv files
cleaned_jan= pd.read_csv("../tableau_challenge/cleaned_jan_citi_bike.csv")
cleaned_jan.head()
# print(len(cleaned_jan))
```

```python
#tried to merge on tableau, but not successful therefore i concatenated the two and double checked the length to make sure no rows were deleted
concat_df = pd.concat([cleaned_dec, cleaned_jan], ignore_index=True)
concat_df.head()
len(concat_df)
```

```python
#saving final as final csv
concat_df.to_csv('final_citi_bike_data.csv', index=False)
```
